**Adding new features to the existing dataset**

```
df = df.withColumn("Alcohol_to_Bitterness", col("Alcohol_Content") / col("Bitterness"))
df = df.withColumn("Ingredient_Efficiency", col("Volume_Produced") / col("Ingredient_Ratio"))
```

```
from pyspark.sql.functions import concat_ws
df = df.withColumn("SKU_Location", concat_ws("_", col("SKU"), col("Location")))
```

```
df = df.withColumn("Total_Loss", col("Loss_During_Brewing") + col("Loss_During_Fermentation") + col("Loss_During_Bottling_Kegging"))
df = df.withColumn("Loss_Percentage", (col("Total_Loss") / col("Volume_Produced")) * 100)
```

```
df = df.withColumn("Quality_Category", when(col("Quality_Score") >= 8, "High").when((col("Quality_Score") >= 6) & (col("Quality_Score") < 8), "Medium").otherwise("Low"))
```

```
df = df.withColumn("Sales_Efficiency", col("Total_Sales") / col("Volume_Produced"))
```

```
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline
```

```
num_col =[
        "Alcohol_Content", "Bitterness", "pH_Level", "Fermentation_Time",
        "Temperature", "Sales_Efficiency", "Loss_Percentage"
    ]
for col_name in num_col:
    null_count = df.filter(col(col_name).isNull()).count()
    print(f"{col_name} has {null_count} null values.")

df = df.fillna(0, subset=num_col) 
```

```
indexer = StringIndexer(inputCol="Quality_Category", outputCol="Quality_Index")
assembler = VectorAssembler(
    inputCols=[
        "Alcohol_Content", "Bitterness", "pH_Level", "Fermentation_Time",
        "Temperature", "Sales_Efficiency", "Loss_Percentage"
    ],
    outputCol="features"
)

dt = DecisionTreeClassifier(labelCol="Quality_Index", featuresCol="features")

pipeline = Pipeline(stages=[indexer, assembler, dt])

train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)
```

```
model = pipeline.fit(train_df)

predictions = model.transform(test_df)
```

```
evaluator = MulticlassClassificationEvaluator(
    labelCol="Quality_Index", predictionCol="prediction", metricName="accuracy"
)
accuracy = evaluator.evaluate(predictions)
print(f"Decision Tree Accuracy: {accuracy:.2f}")
```

```
from pyspark.sql.functions import lit
from graphframes import GraphFrame
```

```
edges = df.select(col("SKU_Location").alias("src"), lit("Network").alias("dst"))
vertices = edges.selectExpr("src as id").union(edges.selectExpr("dst as id")).distinct()

graph = GraphFrame(vertices, edges)
```

```
pagerank = graph.pageRank(resetProbability=0.15, maxIter=10)
pagerank.vertices.show()
```

```
columns_to_analyze = ["Fermentation_Time", "Temperature", "pH_Level", "Gravity", "Alcohol_Content", "Bitterness", "Color", "Quality_Score"]
data_filtered = df.select(*columns_to_analyze)
data_cleaned = data_filtered.dropna()

```

```
vector_col = "features"
assembler = VectorAssembler(inputCols=columns_to_analyze[:-1], outputCol=vector_col)
data_vectorized = assembler.transform(data_cleaned)
```

```
from pyspark.ml.stat import Correlation
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
pearson_matrix = Correlation.corr(data_vectorized, vector_col, "pearson").head()[0].toArray()
corr_df = pd.DataFrame(pearson_matrix, columns=columns_to_analyze[:-1], index=columns_to_analyze[:-1])
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Brewing Parameters")
plt.show()
```

```
from pyspark.sql.functions import col, sum, avg, count, desc
columns_to_analyze = ["Beer_Style", "SKU", "Location", "Total_Sales", "Volume_Produced"]
data_filtered = df.select(*columns_to_analyze)
data_cleaned = data_filtered.dropna()
sales_by_style = data_cleaned.groupBy("Beer_Style").agg(sum("Total_Sales").alias("Total_Sales"),avg("Total_Sales").alias("Average_Sales"),count("Beer_Style").alias("Number_of_Batches")).orderBy(desc("Total_Sales"))
sales_by_location = data_cleaned.groupBy("Location").agg(sum("Total_Sales").alias("Total_Sales"),avg("Total_Sales").alias("Average_Sales"),count("Location").alias("Number_of_Batches")).orderBy(desc("Total_Sales"))


                                                         


```

```
sales_by_sku = data_cleaned.groupBy("SKU").agg(sum("Total_Sales").alias("Total_Sales"),avg("Total_Sales").alias("Average_Sales"),count("SKU").alias("Number_of_Batches")).orderBy(desc("Total_Sales"))
sales_by_sku.show()
```

```
efficiency_df = df.select("Brewhouse_Efficiency", "Loss_During_Brewing", "Loss_During_Fermentation", "Loss_During_Bottling_Kegging","Beer_Style", "Location")
efficiency_df = efficiency_df.select(*(col(c).cast("double").alias(c) if c != "Beer_Style" and c != "Location" else col(c) for c in efficiency_df.columns))
```

```
efficiency_metrics = efficiency_df.groupBy("Beer_Style").agg(avg("Brewhouse_Efficiency").alias("Avg_Brewhouse_Efficiency"),avg("Loss_During_Brewing").alias("Avg_Loss_Brewing"),avg("Loss_During_Fermentation").alias("Avg_Loss_Fermentation"),avg("Loss_During_Bottling_Kegging").alias("Avg_Loss_Bottling_Kegging")
).orderBy(desc("Avg_Brewhouse_Efficiency"))
efficiency_metrics.show()
```

```
efficiency_metrics_pd = efficiency_metrics.toPandas()
plt.style.use('ggplot')
plt.figure(figsize=(12, 6))
plt.bar(efficiency_metrics_pd["Beer_Style"], efficiency_metrics_pd["Avg_Brewhouse_Efficiency"], color='green', alpha=0.7,log = True)
plt.title("Average Brewhouse Efficiency by Beer Style", fontsize=14)
plt.xlabel("Beer Style", fontsize=12)
plt.ylabel("Brewhouse Efficiency (%)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
